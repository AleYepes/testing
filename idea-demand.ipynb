```python
import pandas as pd                        
from pytrends.request import TrendReq
from time import sleep
from pytrends import exceptions

pytrend = TrendReq()

def sortServices(services):
    try:
        sorting = pd.Series()
        comparison_set = []
        for i, service in enumerate(services):
            if len(comparison_set) == 5:
                sleep(2)

                pytrend.build_payload(kw_list=comparison_set, timeframe='today 5-y')
                relative_demand = pytrend.interest_over_time().sum().sort_values(ascending=False)

                if relative_demand.index[0] == comparison_set[0]:
                    comparison_set = [relative_demand.index[0]]
                    sorting = pd.concat([sorting, relative_demand])
                else:
                    sorting = pd.concat([sorting, relative_demand])
                    services = sorting.sort_values(ascending=False).index.to_list() + services[i+1:]
                    sorting = sortServices(services)
                    break
            else: 
                comparison_set.append(service)

        pytrend.build_payload(kw_list=comparison_set, timeframe='today 5-y')
        relative_demand = pytrend.interest_over_time().sum().sort_values(ascending=False)

        if relative_demand.index[0] == comparison_set[0]:
            sorting = pd.concat([sorting, relative_demand[relative_demand > 1]])
        else:
            sorting = pd.concat([sorting, relative_demand[relative_demand > 1]])
            services = sorting.sort_values(ascending=False).index.to_list() + services[i+1:]
            sorting = sortServices(services)

        return sorting.sort_values(ascending=False)
    except exceptions.TooManyRequestsError:
        print('TooMany')
        return sorting.sort_values(ascending=False)
```
