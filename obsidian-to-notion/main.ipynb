```python
import os

def list_files(startpath, output_file):
    for root, dirs, files in os.walk(startpath):
        level = root.replace(startpath, '').count(os.sep)
        indent = ' ' * 4 * (level)
        output_file.write('{}# {}\n'.format(indent, os.path.basename(root)))
        subindent = ' ' * 4 * (level + 1)
        for f in files:
            filename_without_ext = os.path.splitext(f)[0]
            output_file.write('{}{}\n'.format(subindent, filename_without_ext))

with open('output.md', 'w', encoding='utf-8') as f:
    list_files(r'C:\Users\Alex\Documents\Notes', f)
```
