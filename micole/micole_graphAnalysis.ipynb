```python
def graph2metrics(G):
    nodes = list(G.nodes)
    degrees = []
    in_degrees = []
    out_degrees = []

    # https://cambridge-intelligence.com/keylines-faqs-social-network-analysis/
    # Undirected centrality
    betweenness_centrality = [] #  indicates bridge/bottleneck nodes
    closeness_centrality = [] # average steps to all other nodes. ~Center of mass
    degree_centrality = [] # degree normalized. Relative neighbor count
    eigenvector_centrality = [] # measures neighbors' importance, which is determined by first, second, etc order degrees (in and out)
    clustering_coeff = [] # measures neighbors' local interconnectedness
    # Directed centrality
    pageranks = [] # probability of a random walk reaching a node. Directed eigen: measures neighbors' in_degree

    for node, degree in G.degree():
        degrees.append(degree)
    for node, degree in G.in_degree():
        in_degrees.append(degree)
    for node, degree in G.out_degree():
        out_degrees.append(degree)
    for node, centrality in nx.betweenness_centrality(G).items():
        betweenness_centrality.append(centrality)
    for node, centrality in nx.degree_centrality(G).items():
        degree_centrality.append(centrality)
    for node, centrality in nx.closeness_centrality(G).items():
        closeness_centrality.append(centrality)
    for node, centrality in nx.eigenvector_centrality(G).items():
        eigenvector_centrality.append(centrality)
    for node, pagerank in nx.pagerank(G).items():
        pageranks.append(pagerank)
    for node, coeff in nx.clustering(G).items():
        clustering_coeff.append(coeff)

    data = []
    for i, node in enumerate(nodes):
        data.append([nodes[i], degrees[i], in_degrees[i], out_degrees[i], betweenness_centrality[i], degree_centrality[i], closeness_centrality[i], eigenvector_centrality[i], pageranks[i], clustering_coeff[i]])
    
    df = pd.DataFrame(data, columns=["nodes", "degrees", "in_degrees", "out_degrees", "betweenness_centrality", "degree_centrality", "closeness_centrality", "eigenvector_centrality", "pageranks", "clustering_coeff"])
    df['total_centrality'] = df.betweenness_centrality + df.pageranks
    # df = df.sort_values(by=['total_centrality'], ascending=False)
    return df
```

```python
def calculate_real_potential_ratios(G, source, sink, capacities, real_flows):
  """
  Calculates real/potential ratios for each node in the network.

  Args:
      G: NetworkX graph representing the interaction flow.
      source: Source node of the flow (entering the site).
      sink: Sink node of the flow (purchasing a product).
      capacities: Dictionary containing maximum capacities for each edge.
      real_flows: Dictionary containing actual traffic data for each interaction (node).

  Returns:
      Dictionary with real/potential ratios for each node in the network.
  """
  remaining_flow = capacities[source]  # Start with full potential flow
  print(remaining_flow)
  node_ratios = {}

  while remaining_flow > 0:
    # Calculate maximum flow considering current remaining flow
    flow_value, flow_dict = nx.maximum_flow(G, source, sink, capacity=remaining_flow)

    for node, flow in flow_dict.items():
      if node in real_flows:
        # Calculate real/potential ratio for the node
        potential_flow = capacities.get(node, remaining_flow) 
        node_ratios[node] = real_flows[node] / (potential_flow or 1)  # Avoid division by zero

      # Update remaining flow for next iteration
      remaining_flow -= flow

    # Update capacities for the next iteration based on actual flow
    for u, v, d in G.edges(data=True):
      if u in flow_dict and v in flow_dict:
        capacities[(u, v)] = min(capacities[(u, v)], capacities[u] - flow_dict[u])

  return node_ratios


# Create a directed graph
G = nx.DiGraph()

# Add edges based on the flowchart
G.add_edge('entering', 'A')
G.add_edge('entering', 'B')
G.add_edge('entering', 'C')
G.add_edge('A', 'purchasing')
G.add_edge('B', 'purchasing')
G.add_edge('C', 'purchasing')

# Example usage
real_flows = {'entering': 50,'A': 20, 'B': 10, 'C': 5, 'purchasing': 50}  # Example traffic data
capacities = {'entering': 50,'A': 50, 'B': 50, 'C': 50, 'purchasing': 50}
# ratios = calculate_real_potential_ratios(G, 'entering', 'purchasing', capacities, real_flows)
# print(ratios)

nx.maximum_flow(G, 'entering', 'purchasing', capacity=50)
```
