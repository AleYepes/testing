## Aggregate response data

```python
def parse_question(num):

    answer_cols = ['answer_1', 'answer_2', 'answer_3', 'answer_4', 'answer_5']
    df_question = pd.DataFrame()

    for col in answer_cols:
        df_temp = df[df['question_id'] == num][col].value_counts().reset_index()
        question = f'{df[df['question_id'] == num].question.iloc[0]}'
        df_temp.columns = [question, col]
        if df_question.empty:
            df_question = df_temp
        else:
            df_question = pd.merge(df_question, df_temp, how='outer', on=question)

    df_question.fillna(0, inplace=True)
    return df_question

question_counts = {}
for id in df.question_id.unique():
    question_counts[f'{id}'] = parse_question(id)
```
