```python
import pandas as pd


# 'Informativity', 'Speed', 'Price', 'Reliability', 'Accessibility', 'Convenience', 'Compatibility', 'Transparency', 'Control'

# Assuming df1 is your first DataFrame
users_df = pd.DataFrame({
    'users': ['A','B','C','D','E'],
    'reliability': [1, 2, 3, 4, 5],
    'price': [2, 3, 4, 5, 1],
    'speed': [5, 4, 3, 2, 1]
})

# Calculate the mean of each column in df1
requirements_values = users_df.drop('users', axis=1).mean()

# Create df2 with 'requirement' as the column names from df1 and 'mean' as the calculated means
requirements_df = pd.DataFrame({
    'requirement': requirements_values.index,
    'mean': requirements_values.values
})

requirements_df
```
